```python
import duckdb
import pandas as pd
```

```python
conn = duckdb.connect(database="data/staging/data.duckdb")
data = conn.execute(
    """
        SELECT 
                date_trunc('week', pickup_datetime) AS period,
                COUNT(1) AS num_trips,
                sum(passenger_count) AS passenger_count,
                sum(total_amount) AS total_amount,
                sum(trip_distance) AS trip_distance
            from trips
            GROUP BY date_trunc('week', pickup_datetime)
    """
    ).fetch_df()
```

```python
import duckdb
with duckdb.connect(database="data/staging/data.duckdb") as conn:
    data =conn.execute("""
            SELECT 
                date_trunc('week', pickup_datetime) AS period,
                *
            from trips
                WHERE pickup_datetime >= '2023-01-15'::date
                and pickup_datetime < '2023-01-15'::date + interval '1 week'
                and period = '2023-01-09'
            limit 10
""").fetch_df()

display(data)
```
